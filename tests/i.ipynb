```python
import pandas as pd

# 路径配置（根据你的 log）
split_path = '../../../data/antibio/splits/antibio_hplus_80_0.tsv'
labels_path = '../../../data/antibio/labels.tsv'

try:
    # 读取 Split File (通常第一列是 ID)
    # 注意：如果不确定是否有表头，可能需要调整 header=None 或 names=['spec']
    splits = pd.read_csv(split_path, sep='\t')
    split_ids = set(splits.iloc[:, 0].astype(str)) # 取第一列作为ID
    print(f"Split File 加载 ID 数量: {len(split_ids)}")
    print(f"Split ID 示例: {list(split_ids)[:3]}")

    # 读取 Labels File
    labels = pd.read_csv(labels_path, sep='\t')
    # 确保刚才改好的 spec 列存在
    if 'spec' in labels.columns:
        label_ids = set(labels['spec'].astype(str))
        print(f"Labels File 加载 ID 数量: {len(label_ids)}")
        print(f"Labels ID 示例: {list(label_ids)[:3]}")
        
        # 计算交集
        common = split_ids.intersection(label_ids)
        print(f"\n>>> 匹配到的 ID 数量 (交集): {len(common)}")
        
        if len(common) == 0:
            print("\n[错误原因确认]: 两个文件的 ID 完全不匹配！请修改 Labels 文件或 Split 文件使它们一致。")
    else:
        print("Labels file 中没有找到 'spec' 列。")

except Exception as e:
    print(f"读取出错: {e}")
```
